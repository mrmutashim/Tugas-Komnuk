### Soal 1

Fungsi Bessel jenis pertama didefinisikan sebagai berikut

$$ J_\alpha (x) = \sum_{m=0}^{\infty} \dfrac{(-1)^m}{m! \ \Gamma (m+\alpha + 1)} \left( \dfrac{x}{2} \right)^{2m+\alpha} $$

- Buatlah kode Python yang meminta pengguna untuk memilih akan menggunakan metode Bisection ataukah metode Newton. Kemudian pengguna diminta untuk memasukkan tebakan awal atau rentang awal, tergantung dari metode yang dipilih.
- Dengan menggunakan input tersebut, carilah akar dari $J_0$. Di setiap iterasi tampilkan ke layar nilai dari $J_0(x)$ untuk tebakan sekarang dan perubahan pada tebakannya.
- Untuk pengecekan, terdapat akar pada $x \approx 2,4048$.  Juga fungsi <tt>math.gamma(x)</tt> akan memberikan kepada kita nilai dari $\Gamma(x)$.

*Jawab:*

```python
import math
def bisection(a,b,epsilon=1.0e-6):
    n = math.ceil(math.log2((b-a)/epsilon))
    print(f"Perlu {n} iterasi sampai konvergen")
    i = 1
    while (b-a)/2 > epsilon:
        c = (a+b)/2
        J_a = bessel(0,a)
        J_c = bessel(0,c)
        print(f"Iterasi ke-{i} \n  J_0({a}) = {J_a} \n  J_0({c}) = {J_c} \n  Perubahan pada tebakan = {abs(c-a)}")
        if J_c == 0:
            return c
        elif J_a*J_c < 0:
            b = c
        else:
            a = c
        i += 1
    c = (a+b)/2
    J_a = bessel(0,a)
    J_c = bessel(0,c)
    print(f"Iterasi ke-{n} \n  J_0({a}) = {J_a} \n  J_0({c}) = {J_c} \n  Perubahan pada tebakan = {abs(c-a)}")
    return c

def inexact_newton(x0,delta = 1.0e-7, epsilon=1.0e-6, LOUD=False):
    print(f"x0 = {x0}")
    J_x0delta = bessel(0,x0+delta)
    J_x0 = bessel(0,x0)
    faksen = (J_x0delta-J_x0)/delta
    h = J_x0/faksen
    i = 0
    while abs(h) > epsilon:
        i += 1
        if LOUD == True:
            J_x0delta = bessel(0,x0+delta)
            J_x0 = bessel(0,x0)
            faksen = (J_x0delta-J_x0)/delta
            x = x0-(J_x0/faksen)
            print(f"Iterasi ke-{i} \n  J_0({x0}) = {bessel(0,x0)} \n  J_0({x}) = {bessel(0,x)} \n  Perubahan pada tebakan = {abs(x-x0)}")
        x0 = x
        h = bessel(0,x0)/faksen
    print(f"Perlu {i} iterasi sampai konvergen")
    return x

def bessel(alpha,x,i=100):
    hasil = 0
    for m in range(i):
        J = (((-1)**m)/(math.factorial(m)*math.gamma(m+alpha+1)))*((x/2)**(2*m+alpha))
        hasil += J
    return hasil
```

```python
metode = input("Masukkan pilihan metode (Bisection/Newton) : ")
if metode == "Bisection":
    a = float(input("Masukkan batas bawah rentang awal : "))
    b = float(input("Masukkan batas atas rentang awal : "))
    x = bisection(a,b)
    print(f"Akar dari J_0 adalah {x}")
elif metode == "Newton":
    x0 = float(input("Masukkan tebakan awal : "))
    x = inexact_newton(x0,LOUD=True)
    print(f"Akar dari J_0 adalah {x}")
else:
    print("Masukkan nama metode dengan benar")
```

### Soal 2

Radionuklida $^{135}$Xe dihasilkan di dalam reaktor nuklir melalui dua mekanisme :
- Produksi $^{135}$Te dari reaksi fisi yang kemudian  menjadi $^{135}$I melalui peluruhan $\beta$ dengan waktu paro 19 detik, dan selanjutnya meluruh menjadi $^{135}$Xe melalui perluruhan $\beta$ dengan waktu paro 6,6 jam.
- Produksi $^{135}$Xe langsung dari reaksi fisi.

Nuklida $^{135}$Xe ini penting karena ia mempunyai tampang lintang tangkapan yang sangat besar yaitu $2,6\times 10^6$ barn. $^{135}$Xe juga bersifat radioaktif dan meluruh dengan waktu paro sebesar 9,1 jam.

Untuk reaktor dengan fluks skalar rerata $\phi$ dalam satuan neutron/cm$^2$/s, persamaan yang menyatakan produksi dari $^{135}$Xe adalah

$$
\begin{align}
\frac{d}{dt} T &= \gamma_T \Sigma_f \phi - \lambda_T T \\
\frac{d}{dt} I &= \lambda_T T - \lambda_I I \\
\dfrac{d}{dt} X &= \lambda_I I + \gamma_X \Sigma_f \phi - \left(\lambda_X + \sigma_a^X \phi \right)X
\end{align}
$$

dengan $T$, $I$ dan $X$ melambangkan densitas nuklida (dalam satuan nuklei/cm$^3$) masing-masing untuk $^{135}$Te, $^{135}$I dan $^{135}$Xe. $\lambda_T$, $\lambda_I$ dan $\lambda_X$ merupakan konstanta peluruhan untuk $^{135}$Te, $^{135}$I dan $^{135}$Xe.

Jika kita mempertimbangkan kondisi ajeg (*steady state*), maka $\frac{d}{dt} = 0$, dan kita akan menjumpai persamaan untuk **konsentrasi setimbang** ketiga nuklida tersebut. Kondisi ini dijumpai ketika reaktor nuklir telah dioperasikan cukup lama. Untuk reaktor berbahan bakar $^{235}$U, *fission yield* untuk Te dan Xe adalah $\gamma_T = 0,061$ dan $\gamma_X = 0,003$ dan tampang lintang makroskopik fisi sebesar $\Sigma_f = 0,07136$ cm$^{-1}$.

Setelah reaktor beroperasi cukup lama, reaktor tersebut kemudian dipadamkan (*shutdown*). Konsentrasi $^{135}$Xe akan berubah terhadap waktu dan dapat dinyatakan sebagai berikut

$$ X(t) = X_\infty e^{-\lambda_X t} + \dfrac{\lambda_I}{\lambda_I - \lambda_X} I_\infty \left( e^{-\lambda_X t} - e^{-\lambda_I t} \right) $$

dengan $t$ adalah waktu semenjak *shutdown*.

Konsentrasi $^{135}$Xe  akan meningkat setelah reaktor dipadamkan, namun akan mencapai nilai maksimum pada waktu tertentu dan kemudian konsentrasinya akan berkurang.

***Hitunglah*** :
- **konsentrasi setimbang** ketiga nuklida tersebut ($T_\infty, I_\infty, X_\infty$) pada densitas daya reaktor sebesar 5, 50 dan 100 W/cm$^3$, dengan besarnya energi yang dibangkitkan sebesar 200 MeV/fisi.
- **laju serapan neutron** pada nuklida $^{135}$Xe yang dinyatakan oleh $\sigma_a^X \phi X$.
- **waktu** yang diperlukan setelah reaktor dipadamkan untuk mencapai konsentrasi $^{135}$Xe  yang maksimum dan juga besarnya **konsentrasi** $^{135}$Xe pada waktu tersebut untuk densitas daya reaktor sebesar 5, 50 dan 100 W/cm$^3$.

*Jawab*

Pada kondisi setimbang, $\frac{d}{dt} = 0$ sehingga persamaan yang menyatakan produksi $^{135}Xe$ dapat dituliskan ke dalam matriks sebagai berikut.

$$\begin{pmatrix}
    \lambda_T & 0 & 0 \\
    -\lambda_T & \lambda_I & 0 \\
    0 & -\lambda_I & \lambda_X + \sigma_a^X \phi \\
\end{pmatrix}
\begin{pmatrix}
    T \\
    I \\
    X \\
\end{pmatrix} =
\begin{pmatrix}
    \gamma_T \Sigma_f \phi \\
    0 \\
    \gamma_X \Sigma_f \phi \\
\end{pmatrix}$$


Untuk itu perlu dicari tahu nilai fluks ($\phi$) yang diperoleh dari nilai densitas daya ($P$), tampang lintang makroskopik fisi ($\Sigma_f$), dan energi per fisi ($E$).
$$
\phi = \frac{P}{\Sigma_f E}
$$

Kemudian matriks di atas diselesaikan dengan metode Gauss-Seidel.

Setelah diperoleh konsentrasi setiap nuklida, maka dapat dihitung laju serapan neutron pada nuklida $^{135}Xe$  dengan menggunakan persamaan:
$$
{Laju~Serapan~Neutron} =\sigma_a ^X \phi X
$$

Setelah reaktor bereaksi cukup lama, reaktor kemudian dipadamkan (*shutdown*). Konsentrasi $^{135}Xe$ akan berubah terhadap waktu dan dapat dinyatakan sebagai berikut.
$$ X(t) = X_\infty e^{-\lambda_X t} + \dfrac{\lambda_I}{\lambda_I - \lambda_X} I_\infty \left( e^{-\lambda_X t} - e^{-\lambda_I t} \right) $$

Untuk menghitung waktu ketika $^{135}Xe$ mencapai maksimum maka perlu dicari nilai t ketika turunan dari persamaan $X(t)$ bernilai 0.
Turunan dari $X(t)$ adalah:
$$
X'(t) = -\lambda_X X_\infty e^{-\lambda_X t} + \dfrac {\lambda_I}{\lambda_I-\lambda_X}I_\infty(-\lambda_X e^{-\lambda_Xt}+\lambda_I e^{-\lambda_It})
$$

Untuk mencari akar dari $X'(t)$ digunakan metode Newton.
Untuk itu diperlukan turunan kedua dari fungsi $X'(t)$
$$
X''(t) = \lambda_X^2 X_\infty e^{-\lambda_X t} + \dfrac {\lambda_I}{\lambda_I-\lambda_X}I_\infty(\lambda_X^2 e^{-\lambda_Xt}-\lambda_I^2 e^{-\lambda_It})
$$

Program yang digunakan sebagai berikut

```python
#Variabel
import numpy as np
lamb_T = np.log(2)/19
lamb_I = np.log(2)/(6.6*3600)
lamb_X = np.log(2)/(9.1*3600)
rho    = [5,50,100]
E      = 200 * 1.60217E-13
sig_f  = 0.07136
sig_a  = 2.6E-18
gam_T  = 0.061
gam_X  = 0.003
```

```python
#Fungsi metode Gauss-Seidel

def Gauss_Seidel(A,b):
    import numpy as np
    N = 3
    x0 = np.zeros(N)
    x = x0.copy()
    epsilon = 1e-6
    converged = False
    while not(converged):
        x0 = x.copy()
        for row in range(N):
            x[row]=b[row]
            for column in range(N):
                if column != row:
                    x[row] -= A[row,column]*x[column]
            x[row] /= A[row,row]
        relative_change = np.linalg.norm(x-x0)/np.linalg.norm(x)
        if relative_change < epsilon:
            converged = True
    return x
```

```python
#Fungsi metode Newton
def Newton(f,f_derivative):
    import numpy as np
    x = 0
    epsilon = 1E-5
    fx = f(x)
    while np.fabs(fx) > epsilon :
        fderx = f_derivative(x)
        x = x - fx/fderx
        fx = f(x)
    return x
```

```python
#Fungsi Persamaan Konsentrasi Xenon-135
def f(t):
    return X0*np.exp(-lamb_X*t) + lamb_I/(lamb_I-lamb_X)*I0*(np.exp(-lamb_X*t)-np.exp(-lamb_I*t))

#Fungsi Turunan Pertama Persamaan Konsentrasi Xenon-135
def f_der1(t):
    return -lamb_X*X0*np.exp(-lamb_X*t) + lamb_I/(lamb_I-lamb_X)*I0*(-lamb_X*np.exp(-lamb_X*t)+lamb_I*np.exp(-lamb_I*t))

#Fungsi Turunan Kedua Persamaan Konsentrasi Xenon-135
def f_der2(t):
    return lamb_X**2*X0*np.exp(-lamb_X*t) + lamb_I/(lamb_I-lamb_X)*I0*(lamb_X**2*np.exp(-lamb_X*t)-lamb_I**2*np.exp(-lamb_I*t))
```

```python
#Program
for p in rho:
    print('Untuk densitas daya = ',p,'W/cm^3:')
    print('='*50)
    phi = p/(sig_f*E)
    print('Nilai phi = %.4E'%phi)
    A = np.array([(lamb_T,0,0),(-lamb_T,lamb_I,0),(0,-lamb_I,(lamb_X+(sig_a*phi)))])
    b = np.array([(gam_T*sig_f*phi),0,(gam_X*sig_f*phi)])
    x = Gauss_Seidel(A,b)
    LSN = sig_a*phi*x[2]
    print('Tellurium-135 = %.4E'%x[0],'atom/cm3')
    print('Iodin-135     = %.4E'%x[1],'atom/cm3')
    print('Xenon-135     = %.4E'%x[2],'atom/cm3')
    print('Laju Serapan Neutron = %.4E'%LSN,'atom/cm3/s')
    X0 = x[2]
    I0 = x[1]
    time = Newton(f_der1,f_der2)
    print('Waktu untuk mencapai konsentrasi maksimum Xenon-135 adalah %.6f'%(time/3600),'jam')
    print('Konsentrasi maksimum Xenon-135 adalah %.4E'%f(time),'atom/cm3\n')
```
